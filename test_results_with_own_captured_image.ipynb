```python
paths=["Downloads/images.jpg",
      "Downloads/images1.jpg",
      "Downloads/indian dog.jpg",
      "Downloads/indian_dog2.jpeg",
      "Downloads/indian_dog3.jpeg"]
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt

model = keras.models.load_model("catvdogbyfeature.keras")
```

```python
for path in paths:
    img = keras.utils.load_img(path, target_size=(150,150))
    img = keras.utils.img_to_array(img)
    img=img/255.0;
    test_input=img.reshape((1,150,150,3));
    pred=model.predict(test_input);
    plt.imshow(img)
    conf=pred;
    if conf>0.5:
        prediction="Dog";
    else:
        prediction="Cat";
        conf=1-conf;
    conf=int(conf*100);
    plt.title(f"Pred:{prediction}, Chance: {conf}%")
    plt.axis("off")
    plt.show()
```
